```python
import tensorflow as tf
import keras
import pandas as pd
import numpy as np
from keras.utils import np_utils
from sklearn import preprocessing
from keras.models import Sequential
from keras.layers.core import Dense, Activation
```

```python
train = pd.read_csv("train.csv")
test = pd.read_csv("test.csv")

def loss_table(table): #欠損の確認
  null_val = table.isnull().sum()
  percent = 100 * table.isnull().sum()/len(table)
  loss_table = pd.concat([null_val, percent], axis=1)
  loss_table_ren_columns = loss_table.rename(columns = {0:"欠損", 1:"%"})
  return loss_table_ren_columns

loss_table(train)
```

```python
loss_table(test)
```

```python
train["Age"] = train["Age"].fillna(train["Age"].median())
test["Age"] = test["Age"].fillna(test["Age"].median())
#Ageの欠損埋め

train_n = train.values
test_n = test.values
train_x = []
test_x = []

train_pclass = train_n[:, [2]]
train_sex =  train_n[:, [4]]
train_age = train_n[:, [5]]

test_pclass = test_n[:, [1]]
test_sex = test_n[:, [3]]
test_age = test_n[:, [4]]
#print(type(pclass))

def pclass_vector(table):
  for i in range(table.shape[0]):
    table[i] = table[i]-1
  table = np_utils.to_categorical(table) #pclassの前処理
  return table

def sex_vector(table):
  for i in range(table.shape[0]): #sexの前処理
    if table[i] == "male":
      table[i] = 1
    else:
      table[i] = 0
  return table

def age_vector(table):
  table = preprocessing.scale(table) #ageの前処理
  return table



train_pclass = pclass_vector(train_pclass)
train_sex = sex_vector(train_sex)
train_age = age_vector(train_age)

test_pclass = pclass_vector(test_pclass)
test_sex = sex_vector(test_sex)
test_age = age_vector(test_age)


"""
print(train_pclass[0])
print(train_sex[0])
print(train_age[0])
"""

for i in range(train_sex.shape[0]):
  #print(np.concatenate([train_pclass[i], train_sex[i], train_age[i]], axis=0))
  train_x = np.append(train_x, np.concatenate([train_pclass[i], train_sex[i], train_age[i]], axis=0))

for i in range(test_sex.shape[0]):
  #print(np.concatenate([train_pclass[i], train_sex[i], train_age[i]], axis=0))
  test_x = np.append(test_x, np.concatenate([test_pclass[i], test_sex[i], test_age[i]], axis=0))

train_x = train_x.reshape([train_sex.shape[0],5])
test_x = test_x.reshape([test_sex.shape[0],5])

train_y = train_n[:, [1]]

train_x = train_x.tolist()
train_y = train_y.tolist()
test_x = test_x.tolist()

print(train_x[:5])
print(train_y[:5])
print(test_x[:5])
```

```python
model = Sequential()

model.add(Dense(16, input_shape=(5, )))
model.add(Activation("relu"))

model.add(Dense(1))
model.add(Activation("softmax"))
```

```python
model.compile(optimizer="adam", loss = "categorical_crossentropy", metrics=["accuracy"])
model.fit(train_x, train_y, epochs=50, batch_size=1, verbose=1)
```

```python
print(model.predict_classes(test_x))
```
